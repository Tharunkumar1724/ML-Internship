```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame creation (replace this with your actual DataFrame)
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [10, 20, 15, 25]
}
df = pd.DataFrame(data)

# Set the index to the 'Category' column (optional, for better plotting)
df.set_index('Category', inplace=True)

# Create a bar plot
df.plot(kind='bar')

# Show the plot
plt.title('Bar Plot Example')  # Optional: Add a title
plt.xlabel('Category')          # Optional: Add x-label
plt.ylabel('Values')            # Optional: Add y-label
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Create a box plot
plt.boxplot(data, vert=True, patch_artist=True)
plt.title('Box Plot')
plt.xlabel('Dataset')
plt.ylabel('Value')
plt.xticks([1, 2, 3], ['Std 1', 'Std 2', 'Std 3'])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame creation (replace this with your actual DataFrame)
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [10, 20, 15, 25],'Values1': [15, 30, 20, 30]
}
df = pd.DataFrame(data)
plt.pie(df['Values'],labels=df['Category'])
```

```python
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [10, 20, 15, 25],'Values1': [15, 30, 20, 30]
}
df = pd.DataFrame(data)
plt.bar(df['Category'], df['Values'], label='Values', color='lightblue')
plt.bar(df['Category'], df['Values1'], bottom=df['Values'], label='Values1', color='lightcoral')

# Set the title and labels
plt.title('Stacked Bar Graph')
plt.xlabel('Category')
plt.ylabel('Values')

# Add a legend
plt.legend(title='Legend')

# Show the plot
plt.show()
```

```python
import pandas as pd
df=pd.read_csv('coffee_consumption_india.csv')
df.head()
```

```python
import matplotlib.pyplot as plt

plt.boxplot(df['Monthly Coffee Expense (INR)'])
plt.xlabel('Monthly Coffee Expense (INR)')
plt.title('Box Plot of Monthly Coffee Expense')
plt.show()
```

To find Outliers using the Boxplot:
           Qr=Q3-Q1
           for lower outlier:
               Q1-1.5*Qr
            for upper outlier:
                Q3+1.5*Qr

Statistical Methods
Z-Score: The Z-score indicates how many standard deviations an element is from the mean. A common threshold for identifying outliers is a Z-score of greater than 3 or less than -3.
