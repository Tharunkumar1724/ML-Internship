Normalization:In this approach we will scaledown the value feature from 0 to 1 X(norm)=X-X(min)/X(max)-X(min)

```python
import pandas as pd
df=pd.read_csv('Rain Prediction Train.csv')
df.head()
```

```python
df1=df[['Humidity9am','Humidity3pm','Pressure9am']]
df1
```

```python
df3=df1.iloc[0:10]
df3
```

```python
from sklearn.preprocessing import MinMaxScaler
scaling=MinMaxScaler()

fd=scaling.fit_transform(df3[['Humidity9am', 'Humidity3pm']])
print(fd)
```

Standardizatin(Z_Score) Normalization s=x-mean/std only when it is used our mean=0 and standard deviation =1

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
scaler.fit_transform(df3[['Humidity9am', 'Humidity3pm']])
```
