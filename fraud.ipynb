```python


# Rename columns to meaningful names
df.columns = [
    'record_id',               # Unnamed: 0
    'customer_number',         # NIS_RAD
    'fraud_case_number',       # NUM_FR
    'detection_date',          # F_DETEC
    'fraud_type',              # TIPO(TR.TIP_FRA)
    'fraud_observation',       # OBS_FRA
    'address_reference',       # REF_DIR
    'total_recovered_amount',  # IMP_TOT_REC
    'recovery_number',         # NUM_REC
    'account_status',          # ESTADO(RE.EST_ACT)
    'device_code',             # CODIGO(AP.CO_APARATO)
    'power_capacity_kw',       # POTENCIA
    'quantity',                # CANTIDAD
    'daily_usage_hours'        # HORAS_USO_DIA
]

 
print(df.head())
```

```python
'''Status	Meaning / Possible Category:

1. PAID, PAID UNDER INSTALMENT AGREEMENT, UNDER INSTALMENT AGREEMENT, BILLED, SENT TO CUSTOMER	------------>   Non-Fraud / Normal


2.POTENTIAL WRITE-OFF, MANUAL DEFINITE WRITE-OFF, MANUAL 
POTENTIAL WRITE-OFF, SUSPENDED, SUSPENSION PROCESS DETAINED, FINAL DEMAND LETTER,   ------------------->    Fraud / Risky / Problematic
CANCELLED, CANCELLED DUE TO TRANSFER TO ANOTHER ACCOUNT

3.AUTOMATICALLY COLLECTED STAFF INVOICE, AUTOMATICALLY COLLECTED ZERO AMOUNT     -------------->    Maybe negligible cases or edge scenarios
'''
```

```python
fraud_statuses = [
    'POTENTIAL WRITE-OFF',
    'MANUAL DEFINITE WRITE-OFF',
    'MANUAL POTENTIAL WRITE-OFF',
    'SUSPENDED',
    'SUSPENSION PROCESS DETAINED',
    'FINAL DEMAND LETTER',
    'CANCELLED',
    'CANCELLED DUE TO TRANSFER TO ANOTHER ACCOUNT'
]

df['is_fraud'] = df['account_status'].apply(lambda x: 1 if x in fraud_statuses else 0)
```

```python
from sklearn.preprocessing import LabelEncoder

# List of columns to encode
categorical_cols = ['fraud_type', 'fraud_observation', 'address_reference', 
                    'account_status', 'device_code']

# Initialize the encoder
le = LabelEncoder()

# Apply label encoding to each categorical column
for col in categorical_cols:
    df[col] = le.fit_transform(df[col].astype(str))

# Now your DataFrame has label-encoded values in those columns
print(df.head())
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
scaler = StandardScaler()

# Scale all features except customer number
scaled_features = scaler.fit_transform(df.drop(['customer_number','is_fraud','detection_date'], axis=1))
```

```python
df_scaled=pd.DataFrame(scaled_features,columns=df.drop(['customer_number','is_fraud','detection_date'],axis=1).columns)
```

```python
df_scaled['customer_number']=df['customer_number']
df_scaled['is_fraud']=df['is_fraud']  
df_scaled['detection_date']=df['detection_date']
```

```python
df_scaled.head()
```

```python
df_scaled=df_scaled.fillna(0)
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
import pandas as pd

# Assuming your target column
target_col = 'is_fraud'

# Separate features and target
X = df_scaled.drop(columns=[target_col])
y = df_scaled[target_col]

# If necessary, drop non-informative columns like IDs
X = X.drop(columns=['record_id', 'customer_number', 'fraud_case_number', 'recovery_number','detection_date','account_status'])  # optional, depending on use-case

# Apply SelectKBest with ANOVA F-test
selector = SelectKBest(score_func=f_classif, k='all')  # or choose a value for k (top k features)

# Fit selector to data
selector.fit(X, y)

# Get feature scores
scores_df = pd.DataFrame({
    'Feature': X.columns,
    'F_Score': selector.scores_,
    'p_value': selector.pvalues_
}).sort_values(by='F_Score', ascending=False)

# Display feature scores
print(scores_df)
```
