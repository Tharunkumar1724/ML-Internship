**Merge**

```python
import pandas as pd
DF=pd.DataFrame({'Name':['Tharun','Vikranth','Vijay','Antony'],'Marks':[90,78,34,67]})
DF1=pd.DataFrame({'Name':['Kumar','Kalyan','Tharun','Joy'],'Marks1:':[90,45,45,76]})
print(DF)
print(DF1)
```

```python
Merge=pd.merge(DF,DF1,how='inner')
Merge
```

```python
Merge=pd.merge(DF,DF1,how='outer')
Merge
```

```python
DF=pd.DataFrame({'Name':['Tharun','Vikranth','Vijay','Antony'],'Marks':[90,78,34,67]})
DF1=pd.DataFrame({'Name':['Kumar','Kalyan','Tharun','Joy'],'Marks':[90,45,45,76]})
Merge=pd.merge(DF,DF1,how='outer')
Merge
```

```python
DF=pd.DataFrame({'Name':['Tharun','Vikranth','Vijay','Antony'],'Marks':[90,78,34,67]})
DF1=pd.DataFrame({'Name':['Kumar','Kalyan','Tharun','Joy'],'Marks':[90,45,45,76]})
Merge=pd.merge(DF,DF1,how='inner')
Merge
```

```python
DF=pd.DataFrame({'Name':['Tharun','Vikranth','Vijay','Antony'],'Marks':[90,78,34,67]})
DF1=pd.DataFrame({'Name':['Kumar','Kalyan','Tharun','Joy'],'Marks1':[90,45,45,76]})
Merge=pd.merge(DF,DF1,how='left')
Merge
```

```python
DF=pd.DataFrame({'Name':['Tharun','Vikranth','Vijay','Antony'],'Marks':[90,78,34,67]})
DF1=pd.DataFrame({'Name':['Kumar','Kalyan','Tharun','Joy'],'Marks1':[90,45,45,76]})
Merge=pd.merge(DF1,DF,how='right')
Merge
```

**Concat** 

```python
DF=pd.DataFrame({'Name':['Tharun','Vikranth','Vijay','Antony'],'Marks':[90,78,34,67]})
DF1=pd.DataFrame({'Name':['Kumar','Kalyan','Tharun','Joy'],'Marks1':[90,45,45,76]})
dfg=pd.concat([DF,DF1],axis=0).reset_index(drop=True)
dfg
```

```python
DF=pd.DataFrame({'Name':['Tharun','Vikranth','Vijay','Antony'],'Marks':[90,78,34,67]})
DF1=pd.DataFrame({'Name':['Kumar','Kalyan','Tharun','Joy'],'Marks1':[90,45,45,76]})
dfg=pd.concat([DF,DF1],axis=1).drop_duplicates().reset_index(drop=True)
dfg
```

```python
DF=pd.DataFrame({'Name':['Tharun','Vikranth','Vijay','Antony'],'Marks':[90,78,34,67]})
DF1=pd.DataFrame({'Name':['Kumar','Kalyan','Tharun','Joy'],'Marks1':[90,45,45,76]})
dfg=pd.concat([DF,DF1],axis=1).drop_duplicates()
dfg
```
