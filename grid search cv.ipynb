```python
import pandas as pd
```

```python
data=pd.read_csv('data.csv')
```

```python
data.head()
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
X = data.drop(['id', 'diagnosis', 'Unnamed: 32'], axis = 1)

y = data['diagnosis']
```

```python
X.shape, y.shape
```

```python
# lr = LogisticRegression()
lr = LogisticRegression(max_iter=5000)
```

```python
lr.fit(X,y)
```

```python
lr.score(X,y)
```

```python
params = [
    {
        'penalty': ['l1'],
        'C': np.logspace(-4, 4, 20),
        'solver': ['liblinear', 'saga'],
        'max_iter': [100, 1000, 2500, 5000]
    },
    {
        'penalty': ['l2'],
        'C': np.logspace(-4, 4, 20),
        'solver': ['lbfgs', 'newton-cg', 'liblinear', 'sag', 'saga'],
        'max_iter': [100, 1000, 2500, 5000]
    },
    {
        'penalty': ['elasticnet'],
        'C': np.logspace(-4, 4, 20),
        'solver': ['saga'],
        'max_iter': [100, 1000, 2500, 5000],
        'l1_ratio': [0.5, 0.7, 0.9]  # Required for elasticnet
    },
    {
        'penalty': ['none'],
        'solver': ['lbfgs', 'newton-cg', 'sag', 'saga'],
        'max_iter': [100, 1000, 2500, 5000]
    }
]

# There are many other parameters that we could use... but for nw will start with this.
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf = GridSearchCV(estimator = lr, param_grid = params, scoring = 'accuracy', cv = 3, verbose=True, n_jobs=-1)
```

```python
clf_fit = clf.fit(X,y)
```

```python
clf_fit.best_estimator_
```

```python
clf_fit.score(X,y)
```

```python
# Mean cross-validated score of the best_estimator
clf_fit.best_score_
```
