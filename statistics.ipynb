```python
import pandas as pd
from scipy.stats import skew, kurtosis

# Select numeric columns
numeric_df = df3.select_dtypes(include='number')

# Define interpretation functions
def interpret_skew(value):
    if value > 1:
        return "Highly right-skewed"
    elif value > 0.5:
        return "Moderately right-skewed"
    elif value < -1:
        return "Highly left-skewed"
    elif value < -0.5:
        return "Moderately left-skewed"
    else:
        return "Approximately symmetric"

def interpret_kurtosis(value):
    if value > 3:
        return "Leptokurtic (Heavy tails, more outliers)"
    elif value < -1:
        return "Platykurtic (Light tails, fewer outliers)"
    elif value < 1:
        return "Mesokurtic (Normal-like tails)"
    else:
        return "Slightly Leptokurtic"

# Calculate and print with interpretation
for column in numeric_df.columns:
    s = skew(numeric_df[column], nan_policy='omit')
    k = kurtosis(numeric_df[column], fisher=True, nan_policy='omit')
    skew_status = interpret_skew(s)
    kurt_status = interpret_kurtosis(k)
    
    print(f"\n Column: {column}")
    print(f"   Skewness: {s:.4f} ‚Üí {skew_status}")
    print(f"   Kurtosis: {k:.4f} ‚Üí {kurt_status}")
```

```python
'''What it means:

Pandas uses excess kurtosis (normal distribution = 0):

> 0: Leptokurtic ‚Üí heavy tails (more outliers than normal)

< 0: Platykurtic ‚Üí light tails (fewer outliers)

‚âà 0: close to normal



kewness Value	Interpretation
0	Perfectly symmetrical
-0.5 to +0.5	Approximately symmetrical
<-0.5 to -1.0	Moderately negatively skewed
>+0.5 to +1.0	Moderately positively skewed
<-1.0	Highly negatively skewed
>+1.0	Highly positively skewed





Excess Kurtosis	Kurtosis Value	Interpretation
< -1.0	< 2	Platykurtic (Very flat, light tails)
-1.0 to +1.0	2 to 4	Mesokurtic (Normal-like tails)
> +1.0	> 4	Leptokurtic (Peaked, heavy tails)''''
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

# Assume df3 is your DataFrame and you want CIs for all numerical columns
numerical_cols = df3.select_dtypes(include=np.number).columns

confidence = 0.95
z = stats.norm.ppf(1 - (1 - confidence) / 2)
n = len(df3)  # You had a typo here: should be len(df3)

# Create a list to collect CI results
results = []

for col in numerical_cols:
    mean = df3[col].mean()
    std = df3[col].std(ddof=1)
    margin = z * (std / np.sqrt(n))
    lower = mean - margin
    upper = mean + margin

    results.append({
        'Feature': col,
        'Mean': mean,
        'CI Lower': lower,
        'CI Upper': upper
    })

# Convert to DataFrame once all results are collected
ci_df = pd.DataFrame(results)

# Print or export
print(ci_df)
```

```python
'''Confidence Intervals for Means ‚Äî Explained Simply
A Confidence Interval (CI) gives a range of values that is likely to contain the true population mean based on a sample.'''
```

```python
from scipy.stats import f

# Two features to compare
feature1 = df3['Max_Consumption'].dropna()
feature2 = df3['Avg_Consumption'].dropna()

# Calculate variances
var1 = np.var(feature1, ddof=1)
var2 = np.var(feature2, ddof=1)

# Compute F-statistic
F_stat = var1 / var2

# Degrees of freedom
dfn = len(feature1) - 1  # numerator
dfd = len(feature2) - 1  # denominator

# Get p-value from F-distribution
p_value = 1 - f.cdf(F_stat, dfn, dfd)

print(f"F-statistic: {F_stat}")
print(f"P-value: {p_value}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

# Features to use (based on your list)
features = [
    'PHASE', 'TARIFF', 'POWER_SUSCRIBED', 'TENSION', 'DIVISION',
    'ACTIVITY_CMS', 'REGION', 'AGENCY', 'payment_mode', 'IsActive',
    'Tenure_Days', 'Avg_Consumption', 'Max_Consumption', 'Total_Payment',
    'Spike_Count', 'Has_inspected', 'inspection_count',
    'fraud_found_category', 'SERVICE_NBR'
]

# Target
target = 'Found_fraud'

# Prepare the data
X = df3[features]
y = df3[target]

# Optional: Standardize features (helps with interpretation)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Fit Linear Regression
model = LinearRegression()
model.fit(X_scaled, y)

# Coefficients
coefficients = model.coef_
intercept = model.intercept_

# Combine with feature names
results = pd.DataFrame({
    'Feature': features,
    'Coefficient': coefficients
}).sort_values(by='Coefficient', ascending=False)

# Show results
print(f"Intercept: {intercept}")
print("\nLinear Regression Coefficients:")
print(results)
```
