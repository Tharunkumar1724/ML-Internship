```python
import numpy as np
import pandas as pd
from scipy.stats import lognorm
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    'data_column': np.random.exponential(scale=2, size=1000) })
```

```python
df['log_data'] = np.log(df['data_column'])
```

```python
shape, loc, scale = lognorm.fit(df['data_column'], floc=0)
```

```python
x = np.linspace(0, df['data_column'].max(), 100)
pdf = lognorm.pdf(x, shape, loc=loc, scale=scale)
```

```python
plt.hist(df['data_column'], bins=50, density=True, alpha=0.5, color='b', label='Data Histogram')
plt.plot(x, pdf, 'r-', label='Log-Normal PDF')
plt.title('Log-Normal Distribution Fit')
plt.xlabel('Data Values')
plt.ylabel('Density')
plt.legend()
plt.show()
```
