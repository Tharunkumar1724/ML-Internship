```python
#merging datav frame
import pandas as pd

data1 = {
    'ID': [1, 2, 3, 4],
    'Name': ['Aarav', 'Vivaan', 'Aditya', 'Vihaan'],
    'Age': [25, 30, 35, 40]
}

df1 = pd.DataFrame(data1)

data2 = {
    'ID': [3, 4, 5, 6],
    'City': ['Mumbai', 'Delhi', 'Bangalore', 'Chennai'],
    'Salary': [70000, 80000, 90000, 100000]
}

df2 = pd.DataFrame(data2)

print("DataFrame 1:")
print(df1)
print("\nDataFrame 2:")
print(df2)

merged_inner = pd.merge(df1, df2, on='ID', how='inner')
print("\nMerged DataFrame (Inner Join):")
print(merged_inner)

merged_outer = pd.merge(df1, df2, on='ID', how='outer')
print("\nMerged DataFrame (Outer Join):")
print(merged_outer)
```
